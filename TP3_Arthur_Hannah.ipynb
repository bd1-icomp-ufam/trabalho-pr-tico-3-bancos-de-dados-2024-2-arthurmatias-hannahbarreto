# Tarefa 7

O objetivo desta tarefa é analisar e descrever os sistemas de arquivos **Ext2**, **Ext3**, **ReiserFS** e **XFS**, destacando suas principais características e comparando as vantagens e desvantagens de cada um. A seguir, apresento a descrição detalhada de cada sistema de arquivos, além de uma tabela comparativa.

## 1. **Ext2 (Second Extended File System)**

O **Ext2** foi um dos primeiros sistemas de arquivos a ser amplamente utilizado no Linux. Ele é uma evolução do sistema de arquivos **Ext** e é conhecido por sua simplicidade e eficiência. O Ext2 não possui suporte a journaling, o que significa que, em caso de falhas, pode ocorrer corrupção de dados. No entanto, ele é bastante rápido e amplamente suportado, o que ainda o torna uma opção válida para determinadas situações.

### Características do Ext2:
- **Sem Journaling**: Não há registros de transações, o que pode resultar em perda de dados após falhas inesperadas.
- **Desempenho**: Melhor desempenho em leituras e gravações rápidas em comparação com sistemas de arquivos com journaling.
- **Compatibilidade**: É suportado por praticamente todas as distribuições Linux.
- **Recuperação**: Em caso de falha, a recuperação pode ser mais demorada, pois não há um mecanismo de journaling.

## 2. **Ext3 (Third Extended File System)**

O **Ext3** é uma evolução do Ext2 e foi projetado para ser compatível com ele. A principal diferença é a introdução do **journaling**, que ajuda a prevenir a corrupção de dados em caso de falhas inesperadas. O journaling melhora a integridade dos dados, tornando o Ext3 mais confiável, mas também pode resultar em um pequeno impacto no desempenho.

### Características do Ext3:
- **Journaling**: A introdução de journaling permite que transações sejam registradas antes de serem aplicadas, o que melhora a integridade dos dados.
- **Desempenho**: Embora o desempenho seja um pouco mais lento que o Ext2 devido ao journaling, ainda é eficiente e adequado para a maioria das necessidades.
- **Compatibilidade**: Totalmente compatível com o Ext2, permitindo a conversão simples de Ext2 para Ext3 sem perda de dados.
- **Confiabilidade**: A maior confiabilidade devido ao journaling, que evita a corrupção de dados.

## 3. **ReiserFS**

O **ReiserFS** é um sistema de arquivos desenvolvido para ser eficiente em termos de desempenho e espaço, especialmente para lidar com grandes quantidades de arquivos pequenos. Ele introduz algumas inovações, como a organização de arquivos em uma estrutura de árvore binária, permitindo maior eficiência na busca e no gerenciamento de arquivos. No entanto, devido a problemas de desenvolvimento e manutenção, o ReiserFS perdeu popularidade.

### Características do ReiserFS:
- **Journaling**: ReiserFS também possui journaling, o que melhora a segurança em caso de falhas.
- **Desempenho**: Excelente desempenho em cenários com muitos arquivos pequenos devido à sua estrutura eficiente.
- **Espaço**: Otimiza o uso de espaço em disco, especialmente para arquivos pequenos.
- **Descontinuação**: O ReiserFS não é mais amplamente mantido, o que diminui sua adoção em novos projetos.

## 4. **XFS**

O **XFS** é um sistema de arquivos de alto desempenho que foi inicialmente desenvolvido pela Silicon Graphics para o sistema operacional IRIX e, mais tarde, foi portado para o Linux. O XFS é conhecido por sua capacidade de lidar com grandes volumes de dados e arquivos grandes, sendo utilizado em servidores e sistemas que requerem alta performance.

### Características do XFS:
- **Journaling**: Suporta journaling, garantindo integridade de dados.
- **Desempenho**: Muito bom desempenho, especialmente em sistemas de arquivos grandes ou de grande capacidade.
- **Escalabilidade**: Excelente escalabilidade para sistemas com grandes volumes de dados.
- **Suporte a arquivos grandes**: Suporta grandes arquivos e volumes, tornando-o ideal para grandes bases de dados e sistemas de arquivos de grande escala.

## Tabela Comparativa

| **Característica**                | **Ext2**                | **Ext3**               | **ReiserFS**            | **XFS**                 |
|------------------------------------|-------------------------|------------------------|-------------------------|-------------------------|
| **Journaling**                     | Não                     | Sim                    | Sim                     | Sim                     |
| **Desempenho**                     | Alto (sem journaling)   | Bom (com journaling)   | Excelente em arquivos pequenos | Excelente em grandes volumes de dados |
| **Compatibilidade**                | Alta                    | Alta (compatível com Ext2) | Baixa (não amplamente utilizado) | Alta                    |
| **Recuperação após falha**         | Mais demorada           | Rápida                 | Boa (com journaling)    | Rápida                 |
| **Capacidade de manipular grandes arquivos** | Limitada            | Limitada               | Boa (para arquivos pequenos) | Excelente              |
| **Uso comum**                      | Sistemas pequenos, antigos | Uso geral, servidores pequenos | Sistemas de arquivos com muitos arquivos pequenos | Sistemas com grandes volumes de dados |

## Conclusão

Cada sistema de arquivos tem suas características e é adequado para diferentes tipos de uso. O **Ext2** é útil em cenários onde o desempenho é crucial e onde a confiabilidade não é uma grande preocupação, enquanto o **Ext3** melhora a confiabilidade com o journaling, tornando-se uma escolha mais comum para sistemas gerais. O **ReiserFS** pode ser uma boa escolha para sistemas que lidam com muitos arquivos pequenos, mas devido à sua descontinuação, seu uso é limitado. O **XFS**, por outro lado, se destaca pela sua excelente escalabilidade e capacidade de lidar com grandes volumes de dados, sendo uma ótima opção para servidores e ambientes de alto desempenho.

# Tarefa 22 - versão A

````python
import psycopg2
import time
import threading
from random import choice

def reserva_assento_versao_a():
    conn = psycopg2.connect("dbname=bancodados user=usuario password=senha")
    cursor = conn.cursor()

    try:
        cursor.execute("BEGIN;")

        # Passo 1: Recupera os assentos disponíveis com bloqueio para evitar conflitos de concorrência
        cursor.execute("SELECT num_voo FROM Assentos WHERE disp = TRUE FOR UPDATE;")
        assentos = [row[0] for row in cursor.fetchall()]

        # Passo 2: Escolhe um assento aleatoriamente (simula o cliente escolhendo)
        assento_escolhido = choice(assentos)
        time.sleep(1)

        # Passo 3: Registra a reserva
        cursor.execute("UPDATE Assentos SET disp = FALSE WHERE num_voo = %s;", (assento_escolhido,))
        
        conn.commit()
        print(f"Reserva realizada no assento {assento_escolhido}")

    except Exception as e:
        conn.rollback()
        print(f"Erro na reserva: {e}")

    finally:
        cursor.close()
        conn.close()
````

# Tarefa 22 - versão B

````python
import psycopg2
import time
import threading
from random import choice

def reserva_assento_versao_b():
    conn = psycopg2.connect("dbname=bancodados user=usuario password=senha")
    cursor = conn.cursor()

    try:
        # Transação para o Passo 1
        cursor.execute("BEGIN;")
        cursor.execute("SELECT num_voo FROM Assentos WHERE disp = TRUE FOR UPDATE;")
        assentos = [row[0] for row in cursor.fetchall()]
        conn.commit()

        # Passo 2: Escolha do assento fora da transação
        assento_escolhido = choice(assentos)
        time.sleep(1)

        # Transação para o Passo 3
        cursor.execute("BEGIN;")
        cursor.execute("UPDATE Assentos SET disp = FALSE WHERE num_voo = %s;", (assento_escolhido,))
        conn.commit()
        print(f"Reserva realizada no assento {assento_escolhido}")

    except Exception as e:
        conn.rollback()
        print(f"Erro na reserva: {e}")

    finally:
        cursor.close()
        conn.close()
````

# Tarefa 23

````python
import matplotlib.pyplot as plt

def simula_reserva(versao_func, k):
    threads = []
    inicio = time.time()

    for _ in range(k):
        t = threading.Thread(target=versao_func)
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    fim = time.time()
    return fim - inicio

k_values = [1, 2, 4, 6, 8, 10]
tempos_versao_a = [simula_reserva(reserva_assento_versao_a, k) for k in k_values]
tempos_versao_b = [simula_reserva(reserva_assento_versao_b, k) for k in k_values]

# Plotando o gráfico
def plot_tempo_reserva(k_values, tempos_versao_a, tempos_versao_b):
    plt.plot(k_values, tempos_versao_a, label="Versão A (Transação Única)", marker='o')
    plt.plot(k_values, tempos_versao_b, label="Versão B (Transações Separadas)", marker='s')
    plt.xlabel("Número de Agentes (k)")
    plt.ylabel("Tempo Total de Reserva (s)")
    plt.title("Tempo de Reserva em Função do Número de Agentes")
    plt.legend()
    plt.show()

plot_tempo_reserva(k_values, tempos_versao_a, tempos_versao_b)
````
