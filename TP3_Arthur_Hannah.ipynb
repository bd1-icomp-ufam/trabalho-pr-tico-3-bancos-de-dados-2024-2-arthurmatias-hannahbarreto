```python
Tarefa 22
- Versão A (Transação Única): Executa todos os passos em uma única transação, garantindo atomicidade e consistência.
- Versão B (Transações Separadas): Realiza o passo de seleção fora da transação, o que pode levar
a problemas de concorrência, mas reduz o tempo de bloqueio.
- A versão A é mais segura, enquanto a versão B pode ser mais rápida, mas requer tratamento de concorrência.
```

```python
Estrutura da tabela
CREATE TABLE Assentos (
    num_voo INT PRIMARY KEY,
    disp BOOLEAN DEFAULT TRUE
);
```

```python
#versão A

import psycopg2
import time
import threading
from random import choice

def reserva_assento_versao_a():
    conn = psycopg2.connect("dbname=bancodados user=usuario password=senha")
    cursor = conn.cursor()

    try:
        cursor.execute("BEGIN;")

        # Passo 1: Recupera os assentos disponíveis com bloqueio para evitar conflitos de concorrência
        cursor.execute("SELECT num_voo FROM Assentos WHERE disp = TRUE FOR UPDATE;")
        assentos = [row[0] for row in cursor.fetchall()]

        # Passo 2: Escolhe um assento aleatoriamente
        assento_escolhido = choice(assentos)
        time.sleep(1)
        
        # Passo 3: Registra a reserva
        cursor.execute("UPDATE Assentos SET disp = FALSE WHERE num_voo = %s;", (assento_escolhido,))
        
        # Confirma a transação
        conn.commit()
        print(f"Reserva realizada no assento {assento_escolhido}")

    except Exception as e:
        conn.rollback()
        print(f"Erro na reserva: {e}")

    finally:
        cursor.close()
        conn.close()

```

```python
#versão B
def reserva_assento_versao_b():
    conn = psycopg2.connect("dbname=bancodados user=usuario password=senha")
    cursor = conn.cursor()

    try:
        #Passo 1
        cursor.execute("BEGIN;")
        cursor.execute("SELECT num_voo FROM Assentos WHERE disp = TRUE FOR UPDATE;")
        assentos = [row[0] for row in cursor.fetchall()]
        conn.commit()

        # Passo 2: Escolha do assento fora da transação
        assento_escolhido = choice(assentos)
        time.sleep(1)

        #Passo 3
        cursor.execute("BEGIN;")
        cursor.execute("UPDATE Assentos SET disp = FALSE WHERE num_voo = %s;", (assento_escolhido,))
        conn.commit()
        print(f"Reserva realizada no assento {assento_escolhido}")

    except Exception as e:
        conn.rollback()
        print(f"Erro na reserva: {e}")

    finally:
        cursor.close()
        conn.close()
```

```python
Tarefa 23
- A Versão B (Transações Separadas) é ligeiramente mais rápida devido ao menor tempo de bloqueio, já que o 
Passo 2 (escolha) é executado fora da transação.
- À medida que o número de agentes (k) aumenta, o tempo total diminui em ambas as versões, 
indicando que o sistema se beneficia do paralelismo.
- A diferença entre as versões é mais evidente com valores altos de k, já que a Versão B lida melhor com a alta concorrência.
- Ou seja, o aumento de k melhora o desempenho, mas a versão B tem vantagem em cenários de alta concorrência.
```

```python
import matplotlib.pyplot as plt

def simula_reserva(versao_func, k):
    threads = []
    inicio = time.time()

    for _ in range(k):
        t = threading.Thread(target=versao_func)
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    fim = time.time()
    return fim - inicio


k_values = [1, 2, 4, 6, 8, 10]
tempos_versao_a = [simula_reserva(reserva_assento_versao_a, k) for k in k_values]
tempos_versao_b = [simula_reserva(reserva_assento_versao_b, k) for k in k_values]

# Plotando o gráfico
def plot_tempo_reserva(k_values, tempos_versao_a, tempos_versao_b):
    plt.plot(k_values, tempos_versao_a, label="Versão A (Transação Única)", marker='o')
    plt.plot(k_values, tempos_versao_b, label="Versão B (Transações Separadas)", marker='s')
    plt.xlabel("Número de Agentes (k)")
    plt.ylabel("Tempo Total de Reserva (s)")
    plt.title("Tempo de Reserva em Função do Número de Agentes")
    plt.legend()
    plt.show()

plot_tempo_reserva(k_values, tempos_versao_a, tempos_versao_b)
```
